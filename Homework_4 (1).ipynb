### <font color='red'>__Question I__</font>
Create a figure sized 5 inches height and 10 inches width, composed of 3 subplots (a 1 by 3 grid)
- First subplot showing $y(t)=sin(\frac{1}{t})$ for $\frac{\pi}{5}\leq t \leq 5\pi$, line color green and line type dash-dot.
- Second one showing a tangent function from $\frac{\pi}{20}$ to $\frac{\pi}{5}$, line color black, hexagonal markers, dashed line style, and line thickness of 4.
- Third one showing $y(t)=1.5e^{-0.2t}\sin(5t+\frac{\pi}{4})$, plotted in blue color, for the range t=0 to t=10.

Make sure all the plots have grid lines, axes have labels, and there is a super title for the figure. Add code lines to save the figure with dpi=300 and to download the figure from the notebook. (You do __not__ need to submit the downloaded figure as part of your submission.)



```python
# Your code goes here
# defining the vector
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x1=np.linspace(np.pi/5,5*np.pi,100)
x2=np.linspace(np.pi/20,np.pi/5,16)
x3=np.linspace(0,10,1000)
y3 = 1.5 * np.exp(-0.2 * x3) * np.sin(5 * x3 + np.pi/4)

fig=plt.figure(figsize=(14,4))               # defining the matrix of plots
fig.suptitle('QUESTION 1')

fig.add_subplot(1,3,1)
plt.plot(x1,np.sin(1/x1),'g-.',linewidth=3)
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.title('y(t)=sin(1t) for  π5≤t≤5π')

fig.add_subplot(1,3,2)
plt.plot(x2,np.tan(x2), marker='H', color='black', linestyle='--', linewidth=4)
plt.xlabel('X-values')
plt.title('Tangent Function')

fig.add_subplot(1,3,3)
plt.plot(x3, y3,'b',linewidth=3)
plt.xlabel('X-values')
plt.title('y(t)=1.5e−0.2tsin(5t+π4)')

plt.savefig('my_plot.png', dpi=300)
```

<font color='blue'>__(2)__</font> Provide a bubble chart, representing the turbine capacity versus turbine roto diameter with the size of the markers representing the project capacity. Show the Turbine characteristic confidence on your plot as well. Do not forget to use option `labels` to properly name all the variables.

```python
# Your code goes here

turbine_capacity = data['t_cap']  # Turbine capacity in kW
rotor_diameter = data['t_rd']  # Turbine rotor diameter in meters
project_capacity = data['p_cap']  # Project capacity in MW
turbine_confidence = data['t_conf_atr']  # Turbine characteristic confidence (0-3)

plt.figure(figsize=(10, 6))
plt.scatter(rotor_diameter, turbine_capacity, s=project_capacity, c=turbine_confidence, cmap='coolwarm', alpha=0.7)

plt.xlabel('Turbine Rotor Diameter (meters)')
plt.ylabel('Turbine Capacity (kW)')
plt.title('Turbine Capacity vs. Rotor Diameter with Project Capacity and Confidence')
cbar = plt.colorbar()
cbar.set_label('Turbine Characteristic Confidence')
plt.grid(True)
plt.show()
```

<font color='blue'>__(3)__</font> Create a map centered on the US (look up the US latitude and longitude) and represent the locations of the wind turbine projects on the map. Use green circles to represent each project and make sure that the project name is shown as a popup for each project. Save the map as an html file and submit it along with your notebook for this assignment.

```python
# Your code goes here
import folium

us_map = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Loop through the data and add markers for each wind turbine project
for _, row in data.iterrows():
    lat, lon = row['ylat'], row['xlong']
    project_name = row['p_name']
    marker = folium.CircleMarker(
        location=[lat, lon],
        radius=6,
        color='green',
        fill=True,
        fill_color='green',
        fill_opacity=0.7,
        popup=folium.Popup(project_name)
    )
    marker.add_to(us_map)

us_map
us_map.save("wind_turbine_projects_map.html")
```

<font color='blue'>__(4)__</font> For the state of Texas, provide a pie chart showing the project capacity as a percentage of total capacity for each year. Make sure that you use option `labels` to properly name the variables with their units and put the years and percentages inside each slice. (**note**: you first need to define a new dataframe that only includes the data for the state of Texas.)

```python
# Your code goes here
texas_data = data[data['t_state'] == 'TX']
total_capacity_by_year = texas_data.groupby('p_year')['p_cap'].sum()
years = [int(year) for year in total_capacity_by_year.index]

plt.figure(figsize=(8, 8))
colors = plt.cm.Paired(range(len(total_capacity_by_year)))
plt.pie(total_capacity_by_year, labels=years, autopct='%1.1f%%', colors=colors)
plt.title('Project Capacity Percentage by Year (Texas)')
plt.axis('equal')
plt.show()
```
